```python
import math
import numpy as np
import matplotlib.pylab as plt
def pltNormal():
    plt.rcParams['ytick.direction'] = 'in'
    plt.rcParams['xtick.direction'] = 'in'
    plt.rcParams['figure.subplot.bottom'] = 0.15
    plt.rcParams['figure.subplot.left'] = 0.15
    plt.rcParams["font.size"]=8

def axNormal(ax):
    ax.xaxis.set_ticks_position('both')
    ax.yaxis.set_ticks_position('both')
    ax.tick_params(axis='x')
    ax.tick_params(axis='y')

pltNormal()
```

```python
fig, axs = plt.subplots(5,5,figsize=(10,10),sharex=True,sharey=True)
for ax in axs.flat:
    axNormal(ax)

data=np.loadtxt("probePosition2nm.dat")
maxz=np.max(data.T[2])
minz=np.min(data.T[2])
N=25
width=float((maxz-minz)/N)

for i in np.arange(N):    
    center=(i+0.5)*width+minz
    min=center-width*0.5
    max=center+width*0.5

    index=np.where(data.T[2]>min)
    dataPost=data[index]
    index=np.where(dataPost.T[2]<max)
    dataPost=dataPost[index]
    axs.flat[i].scatter(dataPost.T[0],dataPost.T[1],s=1)
    axs.flat[i].set_xlim(-15,15)
    axs.flat[i].set_ylim(-15,15)
    axs.flat[i].set_title(r"{:.1f} $\AA$ $<z<$ {:.1f} $\AA$".format(min,max))
print()
```
